```python
import pandas as pd
import matplotlib.pyplot as plt
import locale
from locale import atof
```

```python
n = 20
```

```python
df0 = pd.read_csv("D:\\c_0.csv".format(n), sep = '\s*;\s*', decimal=',', engine='python')
df5 = pd.read_csv("D:\\c_5.csv".format(n), sep = '\s*;\s*', decimal=',', engine='python')
df10 = pd.read_csv("D:\\c_10.csv".format(n), sep = '\s*;\s*', decimal=',', engine='python')
df15 = pd.read_csv("D:\\c_15.csv".format(n), sep = '\s*;\s*', decimal=',', engine='python')
df20 = pd.read_csv("D:\\c_20.csv".format(n), sep = '\s*;\s*', decimal=',', engine='python')
```

```python
#df["avg_time"].applymap(atof)
#df.head()

plt.figure(figsize=(20,10))
plt.plot(df0["N"], df0["min_time"], marker = 's', label = "mysort insert = 0")
plt.plot(df0["N"], df5["min_time"], marker = 's', label = "mysort insert = 5")
plt.plot(df0["N"], df10["min_time"], marker = 's', label = "mysort insert = 10")
plt.plot(df0["N"], df15["min_time"], marker = 's', label = "mysort insert = 15")
plt.plot(df0["N"], df20["min_time"], marker = 's', label = "mysort insert = 20")
plt.xlabel('$N$')
plt.ylabel('$ms$');
plt.legend(loc = 2)
plt.xlim(100000, 150000)
plt.ylim(30, 55)

plt.title("       insert")
plt.savefig('select.jpg'.format(n), bbox_inches='tight')
```
